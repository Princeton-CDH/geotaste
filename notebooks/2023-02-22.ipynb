```python
cols_with_semis=[]
for col in df:
    semi_counts = Counter(str(x).count(';') for x in df[col])
    if set(semi_counts.keys()) != {0}:
        cols_with_semis.append(col)
        print(col)
        pprint(semi_counts)
cols_with_semis
```

```python
for k,v in dict(df[cols_with_semis].iloc[0]).items():
    if ';' in v:
        print(k, v.count(';'), v, sep='\n', end='\n\n')
```
