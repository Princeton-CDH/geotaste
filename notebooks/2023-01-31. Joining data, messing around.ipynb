```python
# Network?
```

```python
import networkx as nx
```

```python
G=nx.Graph()
df = df_mega_borrows_1922

for person, book, date in zip(df.member_uris, df.item_uri, df.start_date):
    person = person[:-1].split('/')[-1]
    book = book[:-1].split('/')[-1]
    if not G.has_node(person): G.add_node(person, node_type='person')
    if not G.has_node(book): G.add_node(book, node_type='book')
    if not G.has_edge(person, book):
        G.add_edge(person,book,weight=1)#,dates=[date])
    else:
        G.edges[person,book]['weight']+=1
        # G.edges[person,book]['dates'].append(date)

G.order(), G.size()
```

```python
nx.write_graphml(G, 'g1922.graphml')
```
