```python
#
m = get_map()
arrow = folium.RegularPolygonMarker(location=latlon_SCO, fill_color='blue', number_of_sides=3, radius=10).add_to(m)


def get_tooltip_row(row): return row.event_id

def add_paths(df, add_to, **opts):
    last_latlon = None
    for ii,(i,row) in enumerate(df.iterrows()):
        latlon = (row.lat, row.lon)
        mark = folium.Marker(latlon, popup=get_tooltip_row(row)).add_to(add_to)
        if last_latlon:
            pline = folium.PolyLine((last_latlon, latlon), opacity=0, **opts).add_to(add_to)
            pline_path = folium.plugins.PolyLineTextPath(pline, '➤', repeat=True, attributes={'fill': opts['color'], 'font-size': '18'}).add_to(add_to)

        last_latlon = latlon
        # if ii>3: break

add_paths(figdf[figdf.book_id=='joyce-portrait-artist-young'], m, color='green')

# add_paths(figdf[figdf.book_id=='woolf-mrs-dalloway'], m, color='blue')

# add_paths(figdf[figdf.book_id=='woolf-mrs-dalloway'], m, color='blue')
add_paths(figdf[figdf.book_id=='christie-murder-mesopotamia'], m, color='red')


m
```
